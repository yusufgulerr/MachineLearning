```python
import matplotlib.pyplot as plt

# X ve Y koordinatlar
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Grafii izme
plt.plot(x, y)

# Grafii ekranda gsterme
plt.show()
# X koordinatlar
x = [1, 2, 3, 4, 5]

# Y koordinatlar
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Grafii izme
plt.plot(x, y1, label="izgi 1")
plt.plot(x, y2, label="izgi 2")

# Grafie balk ve eksen etiketleri ekleme
plt.title("Birden Fazla izgi Grafii")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafiin gsterilecek ksmn belirleme
plt.xlim(1, 5)
plt.ylim(0, 12)

# Grafiin yerleimi
plt.legend()

# Grafii ekranda gsterme
plt.show()
```
