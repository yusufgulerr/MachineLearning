```python
import matplotlib.pyplot as plt

# X ve Y koordinatları
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Grafiği çizme
plt.plot(x, y)

# Grafiği ekranda gösterme
plt.show()
# X koordinatları
x = [1, 2, 3, 4, 5]

# Y koordinatları
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Grafiği çizme
plt.plot(x, y1, label="Çizgi 1")
plt.plot(x, y2, label="Çizgi 2")

# Grafiğe başlık ve eksen etiketleri ekleme
plt.title("Birden Fazla Çizgi Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafiğin gösterilecek kısmını belirleme
plt.xlim(1, 5)
plt.ylim(0, 12)

# Grafiğin yerleşimi
plt.legend()

# Grafiği ekranda gösterme
plt.show()
```
